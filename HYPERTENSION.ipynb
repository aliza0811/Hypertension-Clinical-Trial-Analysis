```python
columns_to_drop = ['Acronym', 'Study URL', 'Other IDs', 'Collaborators', 'Primary Outcome Measures', 'Secondary Outcome Measures', 'Other Outcome Measures', 'Primary Completion Date','First Posted','Results First Posted','Last Update Posted','Study Documents']
```

```python
df_lean=df.drop(columns=columns_to_drop, errors='ignore')
```

```python
print(f"success! you now have {len(df_lean.columns)} clean.columns.")
df_lean.head()
```

```python
df_lean.isnull().sum()
```

```python
#dropping rows where the critical timeline is missing
df_clean=df_lean.dropna(subset=['Start Date', 'Completion Date']).copy() 
```

```python
#converting data string to actual datetime objects
df_clean['Start Date']=pd.to_datetime(df_clean['Start Date'],errors='coerce')
df_clean['Completion Date']=pd.to_datetime(df_clean['Completion Date'],errors='coerce')
#calculating trial duration
df_clean['Duration_days']=(df_clean['Completion Date']-df_clean['Start Date']).dt.days
```

```python
#fill up the gaps
df_clean['Phases']=df_clean['Phases'].fillna('Not Specified')
df_clean['Interventions']=df_clean['Interventions'].fillna('Not Specified')
df_clean['Study Design']=df_clean['Study Design'].fillna('Not Reported')
df_clean['Enrollment']=df_clean['Enrollment'].fillna(0).astype(int)
```
